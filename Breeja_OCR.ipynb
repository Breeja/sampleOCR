```python
import easyocr
import cv2
from google.colab.patches import cv2_imshow
reader = easyocr.Reader(['en'])
```

```python
reader = easyocr.Reader(['en'])
```

```python
img = cv2.imread(r'/content/sample_ocr.png', 0)
blur = cv2.GaussianBlur(img,(5,5),0)
result = reader.readtext(blur)
hash_map_text = {}
for res in result:
    print(f"Text: {res[1]}, Coordinates: {res[0]}")
    hash_map_text[res[1]] = res[0]
```

```python
# font
font = cv2.FONT_HERSHEY_SIMPLEX

# fontScale
fontScale = 1

# Blue color in BGR
color = (255, 0, 0)

# Line thickness of 2 px
thickness = 2

def draw_bounding(image, x, y,text):
  img_with_bound = cv2.rectangle(img, (x[0], x[1]), (y[0], y[1]), (0, 255, 0), 2)
  img_with_bound= cv2.putText(img_with_bound, text, (x[0], x[1]), font,
                   fontScale, color, thickness, cv2.LINE_AA)
  return img_with_bound
```

```python
for k, v in hash_map_text.items():
  text = k
  x,y = v[0], v [2]
  img = draw_bounding(img, x, y, text)
```
